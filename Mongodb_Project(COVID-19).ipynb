Taking input from the user as the name of the month and displaying aal the acses of india in that Month(For june only till 2 june)

```python
print("Enter the month whose Data is Needed")
print("1.January")
print("2.Feburay")
print("3.March")
print("4.April")
print("5.May")
print("6.June")
choice=int(input("Enter you choice of Month "))
```

```python
if(choice==1):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'01/20$'}}):
        pprint(x)
elif(choice==2):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'02/20$'}}):
        pprint(x)
elif(choice==3):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'03/20$'}}):
        pprint(x)
elif(choice==4):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'04/20$'}}):
        pprint(x)
elif(choice==5):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'05/20$'}}):
        pprint(x)
elif(choice==6):
    for x in COVID19DAILY_REPORT.find({'Date':{'$regex':'06/20$'}}):
        pprint(x)
```

Plotting the graph of each months with corresponding number of cases using matplotlib

```python
jan_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'01/20$'}})#count the number of cases in month of january
feb_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'02/20$'}})
march_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'03/20$'}})
april_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'04/20$'}})
may_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'05/20$'}})
june_count=COVID19DAILY_REPORT.count_documents({'Date':{'$regex':'06/20$'}})
```

```python
X=["January","Febuary",'March','April','May']
Y=[jan_count,feb_count,march_count,april_count,may_count]
```

```python
fig,ax=plt.subplots()
ax.plot(X,Y,marker="o")
ax.set_xlabel("MONTHS")
ax.set_ylabel("NO. OF CASES")
ax.set_title("RATE OF CASES IN MONTH")
```

# UPDATION OF IMFORMATION BY THE GOVERNMENT AUTHORITIES

```python
dou=input("Enter the date in format dd/mm/yy")
print("1.Confirmed")
print("2.Cured")
print("3.Deaths")
choice=int(input("Enter the category"))
state=input("Enter the name of the state :")
```

```python
if(choice==1):
    n=input("Enter the updated confirmed cases : ")
    COVID19DAILY_REPORT.update_one({'$and':[{'Date':dou},{'State/UnionTerritory':state}]},{'$set':{'Confirmed':n}})#Updates the confirmed cases
elif(choice==2):
    n=input("Enter the updated cured number : ")
    COVID19DAILY_REPORT.update_one({'$and':[{'Date':dou},{'State/UnionTerritory':state}]},{'$set':{'Cured':n}})
elif(choice==3):
    n=input("Enter the updated number of deaths : ")
    COVID19DAILY_REPORT.update_one({'$and':[{'Date':dou},{'State/UnionTerritory':state}]},{'$set':{'Deaths':n}})
else:
    print("Enter the vailid choice")
```

# LIST ZONES CATEGORY(RED,YELLOW,GREEN)

```python
print("1. Red Zone States")
print("2. Yellow Zone States")
print("3. Green Zone States")
date=input("Enter the date in format dd/mm/yy")
choice=int(input("Enter your choice"))
```

```python
if(choice==1):
    print("The states in Red zone")
    for x in COVID19DAILY_REPORT.aggregate([{'$match':{'$and':[{'Confirmed':{'$gte':5000}},{'Cured':{'$gt':600}},{'Date':date}]}}]):
        print(x)
elif(choice==2):
    print("The states in Yellow zone")
    for x in COVID19DAILY_REPORT.aggregate([{'$match':{'$and':[{'Confirmed':{'$gt':3000,'$lt':5000}},{'Cured':{'$gt':800}},{'Date':date}]}}]):
        print(x)
elif(choice==3):
    print("The states in Green zone")
    for x in COVID19DAILY_REPORT.aggregate([{'$match':{'$and':[{'Confirmed':{'$gt':1000,'$lt':3000}},{'Cured':{'$gt':500}},{'Date':date}]}}]):
        print(x)
```
